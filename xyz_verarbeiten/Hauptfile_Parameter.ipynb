```python
import numpy as np
import matplotlib.pyplot as plt
import import_ipynb
import xyzFile_generieren_SW

#Parameter
#GEOMETRIE
l = 1.100
b = 0.036

radius = 0.025 # Radius der Kreise für Auflager
auflager_liste = [(radius,radius),(l-radius,b-radius),(radius,b-radius),(l-radius,radius)]


#MATERIAL (Glas)
E  = 70e9      #Glas ~ 70 GPa Elastizitätsmodul
nu = 0.22
t  = 0.003     #Dicke [m]
rho = 2500   #kg/m^3 Dichte einer Aluminiumplatte

#GITTER-Dichte der zurückgegebenen Matrix
step = 40

SW_matrix = xyzFile_generieren_SW.startwerte_generieren(l,b,t,step)
"""
def startwerte_generieren(l=2,b=1,t=0.004,step=40):

generiert daten für startwerte mithilfe der funktion f(x) = c - a * (x-l/2)²
wobei
c = 0.012    #höhe in der mitte
a = c/((l/2)**2)

und speichert diese in einem *.xyz file mit dem namen 'dicke'm_startwerte.xyz
"""
# STARTWERTE-Matrix
filename_startwerte = f"{t}m_startwerte.xyz"

with open(filename_startwerte, "w") as f:
    for i in range(step):
        for j in range(step):
            x, y, w = SW_matrix[i, j]
            f.write(f"{x:.6f}\t{y:.6f}\t{w:.6e}\n")

#region PLOTTEN
np.set_printoptions(threshold=np.inf)
data1 = np.loadtxt(filename_startwerte)

x1= data1[:,0]
y1= data1[:,1]
z1= data1[:,2]

# Raster erzeugen
X = np.unique(x1)
Y = np.unique(y1)

X, Y = np.meshgrid(X, Y)
Z1=z1.reshape(len(Y), len(X))


fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.plot_surface(X, Y,Z1,color="red")        # startwerte



ax.view_init(elev=10, azim=-75)
plt.show()
#endregion
```
