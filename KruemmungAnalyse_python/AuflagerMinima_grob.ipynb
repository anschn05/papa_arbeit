```python
import numpy as np

np.set_printoptions(threshold=np.inf)

l = 1100
b = 360
t = 3.6

filename_MW = f"{t}mm_messwerte.xyz"
data = np.loadtxt(filename_MW)

x_MW = data[:,0]
y_MW = data[:,1]
z_MW = data[:,2]

step_square = len(z_MW)
print(step_square)

min = z_MW[0]
indexMin = 0
tol = 10e-2

for i in range(step_square):
    if z_MW[i]<min:
        min = z_MW[i]
        indexMin = i

auflagerPunkte = [(0,0)]

for i in range(step_square):
    if z_MW[i]<min+tol:
        auflagerPunkte.append((float(x_MW[i]),float(y_MW[i])))

auflagerPunkte.remove((0,0))
print(auflagerPunkte)
```
