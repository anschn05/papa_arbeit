```python
import import_ipynb
import HEADER_kirchhoff_love_Circles
import numpy as np


#zu setzende Variablen:

#GEOMETRIE
l = 2 #m Länge
b = 1 #m Breite

radius = 0.025 # Radius der Kreise für Auflager
auflager_liste = [(radius,radius),(l-radius,b-radius),(radius,b-radius),(l-radius,radius)]

#MATERIAL (Glas)
E  = 70e9      #Glas ~ 70 GPa Elastizitätsmodul
nu = 0.22
t  = 0.003     #Dicke [m]
rho = 2500   #kg/m^3 Dichte einer Aluminiumplatte

#GITTER-Dichte der zurückgegebenen Matrix
step = 40


gravitation_matrix = HEADER_kirchhoff_love_Circles.kirchhoff_love(auflager_liste,l,b,radius,E,nu,t,rho,step)
"""
def kirchhoff_love(auflager_liste,l=1,b=1,radius=0.025,E=70e9,nu=0.22,t=0.003,rho=2500,step=40):

simuliert und visualisiert das verhalten/die verwerfung einer perfekt ebenen Glasplatte unter Einfluss der Gravitation
Auflager/Dirichletrandbdg sind dabei Kreise mit Mittelpunkte aus 'auflager_liste' und radius, die ausgeschnitten werden
"""
print(gravitation_matrix)

#speichert die Werte in *.xyz-Datei
with open("gravitation.xyz", "w") as f:
    for i in range(step):
        for j in range(step):
            x, y, w = gravitation_matrix[i, j]
            f.write(f"{x:.6f}\t{y:.6f}\t{w:.6e}\n")


#Werte für Ausgangsglasplatte
def f(x,y):
    return 0.15/4 * x**2

x_points = np.linspace(0,l, step)
y_points = np.linspace(0,b, step)
X, Y = np.meshgrid(x_points, y_points)

#speichert (x,y,z) in startwerte_matrix für alle (x,y) in X,Y
startwerte_matrix = np.zeros((step, step),dtype=object)
for i in range(step):
    for j in range(step):
        startwerte_matrix[i, j] = (float(X[i,j]), float(Y[i,j]), f(X[i, j], Y[i, j]))       #hier: Funktion 'f' statt 'gf_w'

print(startwerte_matrix)

with open("startwerte.xyz", "w") as f:
    for i in range(step):
        for j in range(step):
            x, y, w = gravitation_matrix[i, j]
            f.write(f"{x:.6f}\t{y:.6f}\t{w:.6e}\n")


```
