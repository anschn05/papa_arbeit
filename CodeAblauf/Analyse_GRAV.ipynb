```python
import numpy as np
from math import sqrt
import import_ipynb
import xyzFile_GRAVITATION

np.set_printoptions(threshold=np.inf)


def Analyse_Grav(t,l,b,Auflager,radius,step):
    # radius = 100 #l*b/40000
    # step = 40
    # print("Dicke:",t,"\nlxb:",l,"x",b)
    # print("Auflager:",Auflager)

    if Auflager == "Aussen":
        auflager_liste = [(0,0),(l,0),(l,b),(0,b)]
    elif Auflager == "Mitte":
        auflager_liste = [(l/2,b/2)]
    else:
        raise ValueError("Auflager ist nicht als 'Mitte' oder 'Aussen' definiert!!!")

    filename_GRAV = xyzFile_GRAVITATION.gravitationsEinfluss_Auflager(l,b,t,step,auflager_liste,radius)
    data_GRAV = np.loadtxt(filename_GRAV)

    x_GRAV = data_GRAV[:,0]
    y_GRAV = data_GRAV[:,1]
    z_GRAV = data_GRAV[:,2]

    Min = z_GRAV[0]       
    Max = z_GRAV[0]         # sollte egl 0 werden
    for i in range(step*step):
        if z_GRAV[i] > Max:
            Max = z_GRAV[i]
        if z_GRAV[i] < Min:
            Min = z_GRAV[i]
        continue

    # print("Max:",Max,"\nMin:",Min)
    MaxDifferenz = Max - Min

    # print("Differenz: ",MaxDifferenz)
    return MaxDifferenz

if __name__ == "__main__":
    Radius = 50
    step = 40

    Dicke = [3,4,5,6,8,10,12]
    l_vals = [1000,2000,3000,4000,5000,6000]
    b_vals = [1000,2000,3000]
    Auflager = ['Mitte','Aussen']

    # Dicke = [3]
    # l_vals = [1000] 
    # b_vals = [1000] 
    # Auflager = ['Mitte']   

    # for t in Dicke:
    #     for l in l_vals:
    #         for b in b_vals:
    #             for auf in Auflager:
    #                 Analyse_Grav(t,l,b,auf)

    with open("latexFiles/Tabelle_GRAV.tex", "w") as f:
        f.write("\\documentclass{article}\n\\usepackage[utf8]{inputenc}\n")
        f.write("\\usepackage[T1]{fontenc}\n")
        f.write("\\usepackage{graphicx}\n")
        f.write("\\usepackage[a4paper, margin=1.5cm]{geometry}\n\n")
        f.write("\\begin{document}\n\n")
        f.write("\\renewcommand{\\arraystretch}{1.7}\n")
        f.write("\\setlength{\\tabcolsep}{15pt}\n\n")



        # für jedes Auflager eigene Tabelle
        for auf in Auflager:
            f.write("\\begin{table}[ht]\n\\centering\n")

            f.write("\\textbf{Auflager: %s, Radius: %d mm} \\\\\n" % (auf,Radius))

            f.write("\\resizebox{\\textwidth}{!}{\n")
            # Tabellenkopf
            f.write("\t\\begin{tabular}{|c|" + "c|"*len(Dicke) + "}\n\t\\hline\n")
            f.write("\tlxb ")

            for t in Dicke:
                f.write("& %d mm " % t)
            f.write("\\\\ \\hline\n\t")

            # Zeilen
            for l in l_vals:
                for b in b_vals:

                    f.write("%dx%d " % (l,b))

                    for t in Dicke:
                        print(t,l,b)
                        val = Analyse_Grav(t, l, b, auf,Radius,step)
                        f.write("& %.3f " % val)

                    f.write("\\\\ \\hline\n\t")

            f.write("\\end{tabular}\n}\n\n")
            f.write("\n\\newpage\n\n")
            f.write("\\end{table}\n\\newpage")

        f.write("\n\n\\end{document}")
```
